```python
station = ["A","B","C","D","E","F","G","H","I","J"]
base_time = [2,3,1,0.5,2,0.2,3,4,5,2]     
arrival_time = [0.1, 1.2, 0.5, 0.3, 0.8, 0.83, 2, 3, 4, 1] 

actual_time = []

for i in range(len(station)):
    
    t = base_time[i] - arrival_time[i]
    if t>10/60:
        actual_time.append(t)


min_index = actual_time.index(min(actual_time))



print("Best Station:", station[min_index])
print("Final Charging Time:", actual_time[min_index])
```
